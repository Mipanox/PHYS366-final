# Loading Data (for testing)

```python
import os
from astropy.utils.data import get_readable_fileobj
from astropy.io import fits
from astropy import wcs
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_pol(pol):
    with get_readable_fileobj(pol, cache=True) as e:
        fitsfile = fits.open(e)
        po       = fitsfile[0].data
        header   = fitsfile[0].header
    return po
```

```python
n1333_Q_f = '../../../Data/jcmt_gdpol/scupollegacy_ngc1333_Q.fits'
n1333_U_f = '../../../Data/jcmt_gdpol/scupollegacy_ngc1333_U.fits'
n1333_pa_f = '../../../Data/jcmt_gdpol/scupollegacy_ngc1333_pa.fits'
n2068_Q_f = '../../../Data/jcmt_gdpol/scupollegacy_n2068_Q.fits'
n2068_U_f = '../../../Data/jcmt_gdpol/scupollegacy_n2068_U.fits'
n2068_pa_f = '../../../Data/jcmt_gdpol/scupollegacy_n2068_pa.fits'

n2068_vg_f = '../../../Data/n2068_fegd.fits'
```

```python
n1333_Q = get_pol(n1333_Q_f)[0]
n1333_U = get_pol(n1333_U_f)[0]
n1333_pa = get_pol(n1333_pa_f)
n2068_Q = get_pol(n2068_Q_f)[0]
n2068_U = get_pol(n2068_U_f)[0]
n2068_pa = get_pol(n2068_pa_f)

n2068_vg = get_pol(n2068_vg_f)
n2068_vg = n2068_vg[1:-1,1:-1]
```

### directionless unit length field configurations

```python
def plot_quiver(pa,s=40,figsize=(16,9)):
    plt.figure(figsize=figsize)
    X, Y = np.meshgrid(np.arange(pa.shape[1]),np.arange(pa.shape[0]))
    vx = np.cos(np.deg2rad(pa))
    vy = np.sin(np.deg2rad(pa))
    
    plt.quiver(X, Y, vy, vx, scale=s, headwidth=0)
    
    return vx,vy
```

```python
n2068_px, n2068_py = plot_quiver(n2068_pa)
```

```python
n2068_vx, n2068_vy = plot_quiver(n2068_vg)
```
